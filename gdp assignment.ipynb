The percentage contribution of primary, secondary and tertiary sectors as a percentage of total GDP for all the states

```python
sectors= ['Primary','Secondary','Tertiary','Gross State Domestic Product']
df_Sectors=df_States[df_States.Item.isin(sectors)]
```

```python
#Contribution of sectors as a percentage of total GDP for all the states

States= list(df_Sectors['State'].unique())
percent_Primary=[]
percent_Secondary=[]
percent_Tertiary=[]
GSDP_state=[]

for state in States:
   
    df_temp=df_Sectors[df_Sectors.State==state]
    gsdp=df_temp.loc[df_temp['Item']=='Gross State Domestic Product','2014-15']
    primary= df_temp.loc[df_temp['Item']=='Primary','2014-15']
    secondary=df_temp.loc[df_temp['Item']=='Secondary','2014-15']
    tertiary=df_temp.loc[df_temp['Item']=='Tertiary','2014-15']
    percent_Primary.append(float(primary)/float(gsdp)*100)
    percent_Secondary.append(float(secondary)/float(gsdp)*100)
    percent_Tertiary.append(float(tertiary)/float(gsdp)*100)
    
df_Sector_stacked=pd.DataFrame()
df_Sector_stacked['State']=df_Sectors['State'].unique()
df_Sector_stacked['Primary']=percent_Primary
df_Sector_stacked['Secondary']=percent_Secondary
df_Sector_stacked['Tertiary']=percent_Tertiary  
```

```python
#Visualisation Percentage contribution of sectors as a percentage of total GDP for all the states

#Stacked Bar Chart for Percentage contribution of sectors as a percentage of total GDP for all the states   
   
df_Sector_stacked.plot(x='State',kind='bar',stacked=True,fontsize=18,figsize=(20,12))
plt.xlabel('States',size=26)
plt.ylabel('Percentage contribution of sectors',size=26)
plt.title('Percentage contribution of sectors as a percentage of total GDP for all the states',size=28)

    
```

#Categorise the states into four categories based on GDP per capita (C1, C2, C3, C4 - C1 would have the highest per capita GDP, C4 the lowest)

```python
#Quantile values of GDP per capita for 0.20,0.5,0.85,1

df_GDP_STATES['2014-15'].quantile([0.20,0.5,0.85,1])
```

```python
#Categorising STATES into C1,C2,C3,C4

for state in States:
    x= df_GDP_STATES.loc[df_States['State']==state,'2014-15']
    c1value=df_GDP_STATES['2014-15'].quantile([1])
    c2value=df_GDP_STATES['2014-15'].quantile([0.85])
    c3value=df_GDP_STATES['2014-15'].quantile([0.5])
    c4value=df_GDP_STATES['2014-15'].quantile([0.20])

    if float(c1value) >= float(x) > float(c2value):
         df_States.loc[df_States['State']==state,'Category'] ='C1'
    elif float(c2value) >= float(x) > float(c3value): 
         df_States.loc[df_States['State']==state,'Category'] ='C2'
    elif float(c3value) >= float(x) > float(c4value):
         df_States.loc[df_States['State']==state,'Category'] ='C3'  
    else:
         df_States.loc[df_States['State']==state,'Category'] ='C4' 
                  
```

```python
#datacorrection
df_States.loc[df_States['Item']=='Trade & repair services*','Item']='Trade & repair services'
df_States.loc[df_States['Item']=='Road transport**','Item']='Road transport'
df_States.loc[df_States['Item']=='Road transport*','Item']='Road transport'
df_States.loc[df_States['Item']=='Services incidental to transport*','Item']='Services incidental to transport'


```

Identifying most contributing SUB-SECTORS towards GSDP

```python
#SUB-SECTOR wise data

#items to be avoided
to_avoid_list=['Primary','Secondary','Tertiary','TOTAL GSVA at basic prices', 'Taxes on Products', 'Subsidies on products',"Population ('00)",'Per Capita GSDP (Rs.)']

df_sub= df_States[~df_States.Item.isin(to_avoid_list)]
#list1=list(df_States.Item.unique())

table=df_sub.pivot_table(values='2014-15',index=['Category','Item'],aggfunc=np.sum)

df_Sub_Sectors=table.reset_index()

```

```python

#The top 3/4/5 sub-sectors  which contribute to approx. 80% of the GSDP of each category
#Sorting based on sub-sectors values for 2014-15 and taking cumulative sum of items

#Category C1
C1_table=df_Sub_Sectors[df_Sub_Sectors['Category']=='C1']
gsdp1=C1_table.loc[C1_table['Item']=="Gross State Domestic Product",'2014-15']
C1_table= C1_table.drop(C1_table.index[C1_table['Item']=="Gross State Domestic Product"])

#Sorting on Item values for 2014-15 
C1_table=C1_table.sort_values(by=['2014-15'], ascending=False)
#Item percentage
C1_table['Item%']=C1_table['2014-15'].apply(lambda x: round(float(x)/float(gsdp1)*100))
#Cumulative SUM
C1_table['CumSum']=C1_table['2014-15'].cumsum(axis=0,skipna=True)  
#Cumsum Percentage
C1_table['Sum%']=C1_table['CumSum'].apply(lambda x: round(float(x)/float(gsdp1)*100))  

#Category C2
C2_table=df_Sub_Sectors[df_Sub_Sectors['Category']=='C2']
gsdp2=C2_table.loc[C2_table['Item']=="Gross State Domestic Product",'2014-15']
C2_table= C2_table.drop(C2_table.index[C2_table['Item']=="Gross State Domestic Product"])
#Sorting on Item values for 2014-15
C2_table=C2_table.sort_values(by=['2014-15'], ascending=False)
#Item percentage
C2_table['Item%']=C2_table['2014-15'].apply(lambda x: round(float(x)/float(gsdp2)*100))
#Cumulative SUM
C2_table['CumSum']=C2_table['2014-15'].cumsum(axis=0,skipna=True)
#Cumsum Percentage
C2_table['Sum%']=C2_table['CumSum'].apply(lambda x: round(float(x)/float(gsdp2)*100))

#Category C3
C3_table=df_Sub_Sectors[df_Sub_Sectors['Category']=='C3']
gsdp3=C3_table.loc[C3_table['Item']=="Gross State Domestic Product",'2014-15']
C3_table= C3_table.drop(C3_table.index[C3_table['Item']=="Gross State Domestic Product"])
#Sorting on Item values for 2014-15
C3_table=C3_table.sort_values(by=['2014-15'], ascending=False)
#Item percentage
C3_table['Item%']=C3_table['2014-15'].apply(lambda x: round(float(x)/float(gsdp3)*100))
#Cumulative SUM
C3_table['CumSum']=C3_table['2014-15'].cumsum(axis=0,skipna=True)
#Cumsum Percentage
C3_table['Sum%']=C3_table['CumSum'].apply(lambda x: round(float(x)/float(gsdp3)*100))

#Category C4
C4_table=df_Sub_Sectors[df_Sub_Sectors['Category']=='C4']
gsdp4=C4_table.loc[C4_table['Item']=="Gross State Domestic Product",'2014-15']
C4_table= C4_table.drop(C4_table.index[C4_table['Item']=="Gross State Domestic Product"])
#Sorting on Item values for 2014-15
C4_table=C4_table.sort_values(by=['2014-15'], ascending=False)
#Item percentage
C4_table['Item%']=C4_table['2014-15'].apply(lambda x: round(float(x)/float(gsdp4)*100))
#Cumulative SUM
C4_table['CumSum']=C4_table['2014-15'].cumsum(axis=0,skipna=True)
#Cumsum Percentage
C4_table['Sum%']=C4_table['CumSum'].apply(lambda x: round(float(x)/float(gsdp4)*100))

```

```python
#Education Dropout Datframe

df_education_orig=pd.read_csv("./rs_session243_au570_1.1.csv")
df_education=df_education_orig.rename(index=str, columns={"Level of Education - State": "State"})
df_education=df_education[['State','Primary - 2014-2015.1','Upper Primary - 2014-2015',
                                             'Secondary - 2014-2015','Senior Secondary - 2014-2015']]
df_education=df_education.rename(index=str, columns={"Primary - 2014-2015.1": "Primary - 2014-2015"})

```

```python
#Merge GDP data and education Dropout for analysis

#left Join on State value
df_GDP_dropout=pd.merge(df_GDP,df_education,on='State',how='left')

df_GDP_dropout=df_GDP_dropout.rename(index=str, columns={"2014-15": "Per Capita GSDP-2014-2015(Rs.)"})
df_GDP_dropout=df_GDP_dropout.drop(df_GDP_dropout[['Item','index']],axis=1)         
df_GDP_dropout.fillna(0, inplace=True)
```
